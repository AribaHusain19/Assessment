Question: Given a list of temperatures in Fahrenheit, create a list comprehension to 

convert them to Celsius.

Also calculate the average temperature using numpy.

temperatures_f = [32, 68, 75, 82, 56]

# Your code should:

# 1. Convert to Celsius using: (F - 32) * 5/9

# 2. Round to 2 decimal places

# 3. Calculate mean using nump
 

```python
temp_fah =[32, 68, 75, 82, 56]

temp_celsius =[(x-32)*(5/9) for x in temp_fah]
print(temp_celsius)
mean=np.mean(temp_celsius)
np.round(mean, decimals="%.2f")


```
