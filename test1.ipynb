```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'City': ['New York', 'Los Angeles', 'Chicago']
})
merged_df1 = pd.merge(df1,df2, on='ID', how='inner')
#print(merged_df1)

merged_df2 = pd.merge(df1,df2, on='ID', how="left")
#print(merged_df2)
merged_df3 = pd.merge(df1,df2, on='ID', how='right')
print(merged_df3)

merged_df4 = pd.merge(df1,df2, on='ID', how='outer')
print(merged_df4)
```

```python
#concatenate the dataframes

d1 = pd.concat([df1,df2], axis =0)
#print(d1)
d2 = pd.concat([df1,df2], axis=1)
print(d2)
```
